## Plotting from Pandas DataFrames


```python
car = pd.read_csv("21708170-car-sales.csv")
car
```

```python
car["Price"] = car["Price"].str.replace('[\$\,\.]', '')
car
```

```python
type(car["Price"][0])
```

```python
car["Price"] = car["Price"].str[:-2]
```

```python
car
```

```python
car["Sale Date"] = pd.date_range("1/1/2021", periods=len(car))
car
```

```python
car["Total Sales"] = car["Price"].astype(int).cumsum()
```

```python
car
```

```python
# Plot the Total Sales column
car.plot(x="Sale Date", y="Total Sales")
```

```python
# Reassign price column to int
car["Price"] = car["Price"].astype(int)
# Plot scatter with price column as numeric
car.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
