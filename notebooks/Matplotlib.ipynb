```python
import matplotlib.pyplot as plt
plt.style.use(['default'])
```

```python
x = [1,2,3,4]
y = [53,65,70,80]
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="Title",
      xlabel="x-axis",
      ylabel="y-axis")
# ax.tick_params(labelcolor='w', labelsize='medium', width=3)
fig.savefig("plot.png")
```

## Making figures with numpy arrays
* Line plot
* Scatter plot
* Histogram
* Bar plot
* Subplots

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
#Line plot
fig, ax = plt.subplots()
ax.plot(x, 2 ** x);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#From dictionary
sunday_breakfast = {'Jalebi' : 6,
                   'Fafda' : 10,
                   'Chutni/Kadi' : 15}
fig, ax = plt.subplots()
ax.bar(sunday_breakfast.keys(), sunday_breakfast.values());
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5))
ax1.plot(x, x/2)
ax2.scatter(x, np.sin(x))
ax3.bar(sunday_breakfast.keys(), sunday_breakfast.values())
ax4.hist(np.random.randn(1000));
```

## Plotting from pandas DF

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/car-sales.csv")
df
```

```python
df["Price"] = df["Price"].replace('[\$,]', '', regex=True).astype(float)
df.to_csv("car-sales.csv") # Exporting modified csv
```

```python
df = pd.read_csv("car-sales.csv")
df
```

```python
df["Sale Date"] = pd.date_range("29/12/2020", periods=len(df))
df
```

```python
df["Total Sales"] = df["Price"].cumsum()
df
```

```python
df.plot("Sale Date", "Total Sales")
```

```python
df.plot("Make", "Odometer (KM)", kind="bar")
```
